```python
x = np.array([1,2,3,4])
y = np.array([2,3,4,5,6])
x1, y1 = np.meshgrid(x, y)#生成网格，两个坐标
x1
```

```python
plt.scatter(x1,y1,c = 'r')
plt.show()
```

```python
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)

# 生成网格矩阵
X, Y = np.meshgrid(x, y)

# 计算二维函数值
Z = X**2 + Y**2

# 可视化（等高线图）
plt.contourf(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar(label='$z = x^2 + y^2$')
plt.title('二维函数网格计算与可视化')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
