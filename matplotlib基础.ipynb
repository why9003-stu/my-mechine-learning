```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(8,4),dpi=100)
x = np.linspace(1,10,10)
y = np.sin(x)
plt.plot(x,y)#折线图
plt.title("sin")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()#图例
plt.show()
```

```python
x = np.linspace(1,10,100)
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.plot(x,np.sin(x),color = 'r')
plt.subplot(1,2,2)
plt.plot(x,np.cos(x),color = 'g')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 6))  # 2行2列子图，返回axes数组
axes[0,0].plot(x, np.sin(x))  # 第1行1列子图
axes[0,1].plot(x, np.cos(x))  # 第1行2列子图
axes[1,0].plot(x, np.tan(x))  # 第2行1列子图
axes[1,1].plot(x, np.sin(x)*np.cos(x))  # 第2行2列子图
plt.tight_layout()
plt.show()
```

```python
'''#x, y：x 轴和 y 轴数据（y 可为多组，实现多线绘制）。
color/c：线条颜色（如 'red'/'r'、'blue'/'b'）。
linestyle/ls：线型（如 '-' 实线、'--' 虚线、':' 点线）。
marker：标记点样式（如 'o' 圆圈、'*' 星号、's' 正方形）。
label：图例标签（需配合 plt.legend() 使用）。'''

x = np.linspace(1,10,10)
y = 2*x
plt.plot(x,y,c = 'r',ls = '--',label = 'line',marker = '*',markersize = 3,)
plt.legend()#只有有legend label才能显示
plt.title("line")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
'''x, y：散点的坐标。
s：点的大小（可传入数组实现大小渐变）。
c：点的颜色（可传入数组实现颜色渐变）。
alpha：透明度（0~1，避免点重叠）。
cmap：颜色映射表（如 'viridis'、'coolwarm'）。'''

#下面是主要功能，一些特殊的功能可以问ai现学

plt.figure(figsize=(8,4),dpi = 100,facecolor ='b' )
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.randint(1,200,100)
sizes = np.random.randint(1, 100, 100)  # 随机大小
plt.scatter(x,y,s = sizes,c = colors,alpha=0.3,cmap='viridis',label = "color")
plt.xlabel('X Axis', fontsize=12, color='blue')
plt.ylabel('Y Axis', fontsize=12, color='red')
plt.title('color',fontsize = 20,loc = 'right')#right要打引号，loc位置
plt.grid(axis='y', ls='--', alpha=0.3, color='gray')#添加网格线
plt.xlim(0,2)#改变坐标轴的大小
plt.ylim(0,2)
plt.legend(loc = 'upper center',fontsize = 12)#显示label图例
plt.colorbar()  # 显示颜色条
plt.show()
```

```python
import numpy as np
'''plt.contourf 是 Matplotlib 库中用于绘制填充等高线图的函数，
常与 np.meshgrid 配合使用，用于可视化二维标量场（如温度分布、气压场、函数值分布等）。
它会根据数据的数值范围划分不同的等高线层级，
并为每个层级填充不同的颜色，直观展示数据的空间分布规律。'''
# 生成网格数据
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# 计算二维函数值
Z = np.sin(np.sqrt(X**2 + Y**2))

# 绘制填充等高线图
plt.figure(figsize=(8, 3))
# 绘制填充等高线，设置15个层级，指定颜色映射
plt.subplot(1,2,1)
contour = plt.contourf(X, Y, Z, levels=10, cmap='viridis')

# 添加颜色条（显示数值与颜色的对应关系）
plt.colorbar(contour, label='z值')

# 添加普通等高线（黑色线条，突出层级边界）
plt.contour(X, Y, Z, levels=15, colors='k', linewidths=0.5)

# 设置标题和坐标轴标签
plt.title('填充等高线图：$z = \\sin(\\sqrt{x^2 + y^2})$', fontsize=12)
plt.xlabel('x', fontsize=10)
plt.ylabel('y', fontsize=10)




plt.subplot(1,2,2)
# 添加普通等高线（黑色线条，突出层级边界）
plt.contour(X, Y, Z, levels=15, colors='k', linewidths=0.5)
# 设置标题和坐标轴标签
plt.title('填充等高线图：$z = \\sin(\\sqrt{x^2 + y^2})$', fontsize=12)
plt.xlabel('x', fontsize=10)
plt.ylabel('y', fontsize=10)





plt.show()
```
